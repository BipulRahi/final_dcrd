```python
import cv2
import pytesseract
import numpy as np

def preprocess_image(image_path):
    """Preprocess image for better OCR accuracy."""
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image = cv2.GaussianBlur(image, (5, 5), 0)  # Reduce noise
    image = cv2.threshold(image, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    return image

def extract_text(image_path):
    """Extract and format text while keeping punctuation."""
    image = preprocess_image(image_path)

    extracted_text = pytesseract.image_to_string(image, config='--oem 3 --psm 3')

    # Preserve punctuation but clean extra spaces
    formatted_text = " ".join(extracted_text.split())  # Remove multiple spaces

    return formatted_text

def process_document(image_path):
    """Extract formatted text from the entire document image."""
    text = extract_text(image_path)
    return {"Extracted_Text": text}

# Test with an image
image_path = "1.jpg"
output = process_document(image_path)
print("Formatted Extracted Text:\n", output["Extracted_Text"])
```

```python
import fitz  # PyMuPDF
import pdf2image
import pytesseract
from PIL import Image

def extract_text_from_pdf(pdf_path):
    """Automatically detect and extract text from any type of PDF (text-based or scanned)."""
    doc = fitz.open(pdf_path)
    text = ""

    # Step 1: Try to extract text directly
    for page in doc:
        text += page.get_text("text") + "\n"

    # Step 2: If extracted text is too short (means scanned PDF), apply OCR
    if len(text.strip()) < 50:  # If very little or no text found
        images = pdf2image.convert_from_path(pdf_path)  # Convert PDF to images
        text = ""  # Reset text

        for img in images:
            text += pytesseract.image_to_string(img, config="--oem 3 --psm 6") + "\n"

    return text.strip()

# **Test with any PDF**
pdf_text = extract_text_from_pdf("1.pdf")
print("Extracted Text:", pdf_text)
```

```python
#  jokes crack

import spacy

# Load SpaCy's transformer-based model (make sure you have it installed: `python -m spacy download en_core_web_trf`)
nlp = spacy.load("en_core_web_trf")

def extract_entities_spacy(text):
    """
    Extracts entities (ORG, PERSON, and CAUSE) from text using SpaCy.
    It also corrects minor OCR errors in entity recognition.
    """
    doc = nlp(text)
    entities = {"ORG": [], "PERSON": [], "CAUSE": []}

    for ent in doc.ents:
        if ent.label_ == "ORG":
            entities["ORG"].append(ent.text)
        elif ent.label_ in ["PERSON", "PER"]:
            entities["PERSON"].append(ent.text)
        elif "agreement" in ent.text.lower() or "contract" in ent.text.lower():
            entities["CAUSE"].append(ent.text)

    # Remove duplicates
    for key in entities:
        entities[key] = list(set(entities[key]))

    return entities

# **Test with OCR output**
image_text = extract_text("1.jpg")  # Extract text using OCR
entities = extract_entities_spacy(image_text)

print("Extracted Entities (SpaCy):", entities)
```

```python
from transformers import pipeline
import re

# Load the summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def extract_names_and_date(text):
    """Extracts shareholder names and agreement date."""
    names_pattern = r'full name\s*\((.*?)\)'
    date_pattern = r'become effective on date\s*\(.*?(\d{4})\)'

    names = re.findall(names_pattern, text, re.IGNORECASE)
    date_match = re.search(date_pattern, text, re.IGNORECASE)

    date = date_match.group(1) if date_match else "Not Specified"
    return names, date

def chunk_text(text, chunk_size=1000):
    """Splits text into manageable chunks for summarization."""
    words = text.split()
    return [" ".join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size)]

def summarize_text(text):
    """Summarizes text while keeping key details."""
    if not text.strip():
        return "No text extracted for summarization."

    # Extract important entities
    names, date = extract_names_and_date(text)
    shareholders = ", ".join(names) if names else "Shareholders (Not Detected)"

    text_chunks = chunk_text(text)
    summaries = []
    # for chunk in text_chunks:
    #     word_count = len(chunk.split())
    #     max_len = min(word_count, 120)  # Max length should not exceed word count
    #     min_len = max(30, int(0.5 * max_len))  # Min length should be at least 30 words

    #     summary = summarizer(chunk, max_length=max_len, min_length=min_len, do_sample=False)[0]["summary_text"]
    #     summaries.append(summary)
    summaries = [summarizer(chunk, max_length=250, min_length=50, do_sample=False)[0]["summary_text"]
                  for chunk in text_chunks]

    final_summary = f"This Shareholders Agreement, effective from {date}, involves {shareholders}. " + " ".join(summaries)

    return final_summary

# **Test**
image_text = extract_text("1.jpg")  # Extract text from image
summary_result = summarize_text(image_text)

print("🔹 Final Summary:", summary_result)
```

## 🚀 Improved Risk Analysis with Sentiment & Contextual Evaluation
We'll categorize risk levels (Low, Medium, High) using:

Keyword Frequency – Detects legal/risk-related words.

Sentiment Analysis – Checks how negative the sentence is.

Contextual Severity – Determines the seriousness of the issue using a pre-trained model.

## Risk Level Criteria
🔵 Low: Few risk words, neutral/positive sentiment.

🟡 Medium: Some risk words, slightly negative sentiment.

🔴 High: Many risk words, strong negative sentiment + high severity context.

### ✅ Final Plan
Extract text from the document.

Identify risk-related sentences.

Perform sentiment analysis to determine negativity.

Use NLI (Natural Language Inference) model to assess severity.

Combine all factors to assign a final risk level.

###📌 Step-by-Step Approach
Extract risk-related sentences using predefined risk words.

Perform sentiment analysis using a high-accuracy transformer model.

Analyze contextual severity with an advanced Natural Language Inference (NLI) model.

Assign final risk levels (Low, Medium, High) based on:

Keyword frequency

Sentiment score

Contextual severity score

```python
import re
from transformers import pipeline

# Load sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")

# Risk-related words
RISK_WORDS = ["penalty", "liability", "termination", "refund", "damage", "fine", "breach", "obligations", "lawsuit"]

def analyze_risks(text):
    """Extracts risk-related sentences, performs sentiment analysis, and assigns risk levels."""

    # Normalize text (remove extra spaces & lowercase)
    cleaned_text = re.sub(r'\s+', ' ', text).strip().lower()

    # Split into sentences
    sentences = re.split(r'(?<=[.!?])\s+', cleaned_text)

    # Extract risk-related sentences
    risky_sentences = [sent for sent in sentences if any(word in sent for word in RISK_WORDS)]

    # Sentiment Analysis
    sentiment_results = sentiment_analyzer(risky_sentences) if risky_sentences else []

    # Calculate severity based on sentiment
    risk_score = 0
    for result in sentiment_results:
        if result["label"] == "negative":
            risk_score += 2  # High impact
        elif result["label"] == "neutral":
            risk_score += 1  # Medium impact
        # Positive statements don't increase risk

    # Assign risk level based on severity
    if risk_score >= 4:
        risk_level = "🔴 High"
    elif risk_score >= 2:
        risk_level = "🟡 Medium"
    else:
        risk_level = "🔵 Low"

    return {
        "risk_sentences": risky_sentences,
        "sentiment_results": sentiment_results,
        "risk_level": risk_level
    }

# **Test the improved function**
image_text = extract_text("1.jpg")  # Extract text from image
risk_analysis_result = analyze_risks(image_text)

print("🔍 Risk Analysis:", risk_analysis_result)
```

```python
import os
import asyncio
import concurrent.futures
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)  # Ngrok auto-starts when app runs



async def run_parallel_tasks(extracted_text):
    loop = asyncio.get_event_loop()
    with concurrent.futures.ThreadPoolExecutor() as executor:
        tasks = [
            loop.run_in_executor(executor, summarize_text, extracted_text),
            loop.run_in_executor(executor, analyze_risks_improved, extracted_text),
        ]
        summary_result, risk_result = await asyncio.gather(*tasks)

    return summary_result, risk_result


@app.route('/photo', methods=['POST'])
def upload_photo():
    if 'file' not in request.files:
        return jsonify({"error": "No file part"}), 400

    file = request.files['file']
    file_path = "1.jpg"
    file.save(file_path)

    extracted_text = extract_text(file_path, "image")
    os.remove(file_path)

    return jsonify({"Extracted_Text": extracted_text})


@app.route('/pdf', methods=['POST'])
def upload_pdf():
    if 'file' not in request.files:
        return jsonify({"error": "No file part"}), 400

    file = request.files['file']
    file_path = "1.pdf"
    file.save(file_path)

    extracted_text = extract_text(file_path, "pdf")
    os.remove(file_path)

    return jsonify({"Extracted_Text": extracted_text})


@app.route('/process', methods=['POST'])
def process_document():
    data = request.json
    extracted_text = data.get("extracted_text")

    if not extracted_text:
        return jsonify({"error": "Missing extracted_text"}), 400

    summary, risks = asyncio.run(run_parallel_tasks(extracted_text))

    return jsonify({
        "Summary": summary,
        "Risk_Analysis": risks
    })

if __name__ == "__main__":
    app.run()
```
